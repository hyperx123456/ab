```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\Lenovo\OneDrive\Documents\DATA_ANALYTICS\DMV_praticals\DMV_praticals\Datasets\retail_sales_dataset.csv")

# Explore the structure and content of the dataset
print(df.info())
print(df.describe())
print(df.head())
```

```python
df.columns
```

```python
region_sales = df.groupby('Product Category')['Total Amount'].sum().reset_index()
region_sales = region_sales.sort_values(by='Total Amount', ascending=False)
print(region_sales)

top_regions = region_sales.head()
print(top_regions)
```

```python
import matplotlib.pyplot as plt

# Bar plot
plt.figure(figsize=(10,6))
plt.bar(region_sales['Product Category'], region_sales['Total Amount'], color='skyblue')
plt.title('Sales Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(8,8))
plt.pie(region_sales['Total Amount'], labels=region_sales['Product Category'],
        autopct='%1.1f%%')
plt.title('Sales Distribution by Category')
plt.show()
```
